# 3 dataset Merging into 1

```python
reviews_hotels = df1.merge(df, on='hotel_id', how='left')

full_data = reviews_hotels.merge(df2, on='user_id', how='left')

full_data.head()
```

# Effect of different Features on Overall rating by user. 

```python
from sklearn.linear_model import LinearRegression
X = full_data[['score_cleanliness', 'score_comfort', 'score_facilities',
               'score_location', 'score_staff', 'score_value_for_money']]
y = full_data['score_overall']

model = LinearRegression().fit(X, y)

predict = model.predict(X)

Coefficient = pd.Series(model.coef_, index=X.columns, name='Regression Coefficient').sort_values(ascending=False).plot(kind='bar', 
    color='skyblue', figsize=(8,6), title='Importance Features', rot=45)
print(Coefficient)

print("R² Score:", r2_score(y, predict))
```

# Calculating Intercept and Coefficient

```python
print('Intercept:', model.intercept_)
print('Coeffeficient:', model.coef_)
```

# Differentiation between explained variation in data as 'R2' score, and unexplained variation in data as 'MSE'.

```python
from sklearn.metrics import mean_squared_error

y_pred = model.predict(X)
residuals = y - y_pred

print("R² Score:", r2_score(y, y_pred)) #Explained Variation
print("MSE:", mean_squared_error(y, y_pred)) #unexplained variation/Residuals
```
